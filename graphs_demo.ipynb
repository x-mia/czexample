```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('czech_examples_database.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
for label in df['label'].unique():
    subset = df[df['label'] == label]
    plt.scatter(subset['gdex'], subset['llm_score'], label=label)

plt.xlabel('gdex')
plt.ylabel('llm_score')
plt.legend(title='Labels')
plt.grid(True)
# plt.savefig('llm_gdex.png', dpi=150)
plt.show()
```

```python
df.groupby('label')['gdex'].mean().plot(kind='bar', figsize=(8,6), color='skyblue')

plt.xlabel('Labels')
plt.ylabel('GDEX')
plt.grid(axis='y')
# plt.savefig('average_gdex.png', dpi=150)
plt.show()
```

```python
df.groupby('label')['llm_score'].mean().plot(kind='bar', figsize=(8,6), color='skyblue')

plt.xlabel('Labels')
plt.ylabel('GDEX')
plt.grid(axis='y')
# plt.savefig('average_llm.png', dpi=150)
plt.show()
```

```python
label_counts = df['label'].value_counts()

label_counts = label_counts.sort_index()

label_counts.plot(kind='bar', figsize=(8,6), color='skyblue')

plt.xlabel('Labels')
plt.ylabel('Count')
plt.grid(axis='y')

plt.xticks(rotation=0)

# plt.savefig('labels_count.png', dpi=150)
plt.show()
```

```python
df1 = df[df['label'] == 'a']
df2 = df[df['label'] == 'b']
df3 = df[df['label'] == 'c']
df4 = df[df['label'] == 'd']
```

```python
plt.plot(df1['gdex'], df1['llm_score'], ".", label='a')
plt.plot(df2['gdex'], df2['llm_score'], ".", label='b')
plt.plot(df3['gdex'], df3['llm_score'], ".", label='c')
plt.plot(df4['gdex'], df4['llm_score'], ".", label='d')
plt.plot([0,14],[0,14])
plt.title('Score Table')
plt.xlabel('GDEX')
plt.ylabel('LLM')
plt.legend()
# plt.savefig('labeled_scores.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(df['label'], df['gdex'], c=df['gdex'], cmap='viridis', s=100, alpha=0.7)
plt.xlabel('Labels', fontsize=12)
plt.ylabel('GDEX', fontsize=12)
plt.colorbar(scatter)

# plt.savefig('scatter_gdex.png', dpi=150)

plt.show()
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(df['label'], df['llm_score'], c=df['llm_score'], cmap='viridis', s=100, alpha=0.7)
plt.xlabel('Labels', fontsize=12)
plt.ylabel('LLM score', fontsize=12)
plt.colorbar(scatter)

# plt.savefig('scatter_gdex.png', dpi=150)

plt.show()
```
